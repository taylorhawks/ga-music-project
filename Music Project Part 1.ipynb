```python
#Libraries
import csv
import pyen
import pandas as pd
import sys
import time
```

```python
#Load EchoNest API
en = pyen.Pyen("YE2YVIWWX827NZMNT")
```

```python
#List of Genres
genre_dict = en.get('genre/list')
full_list_of_genres = [g['name'] for g in genre_dict['genres']]
print full_list_of_genres
```

```python
#There is some weird stuff in there...
#Nintendocore, latvian pop, e6fi, necrogrind
#Let's just look at house music
#(minus "hard house" and "bouncy house")
#House Subgenres
list_of_house_subgenres = ['acid house', 'chicago house', 'dark progressive house', 'deep disco house', 'deep deep house',
                           'deep deep tech house', 'deep euro house', 'deep funk house', 'deep house', 'deep house',
                           'deep progressive house', 'deep soul house', 'deep tech house', 'deep vocal house', 'disco house',
                           'dutch house', 'electro house', 'fidget house', 'filter house', 'greek house', 'hip house', 'house',
                           'lounge house', 'microhouse', 'minimal tech house', 'outsider house']
```

```python
#Table of Artists and their EN IDs, built on top artists for each subgenre
#The structure of this table is Name, ID, Genre

artist_table_rows = []
for house_sg in list_of_house_subgenres:
    sg_artist_dict = en.get('genre/artists', name=house_sg)
    for a in sg_artist_dict['artists']:
        sg_row = (a["name"].encode('utf-8'), a["id"].encode('utf-8'), house_sg)
        artist_table_rows.append(sg_row)
```

```python
#Let's look at the data so far
artist_table_rows[0:20]
```

```python
#Convert to a pandas-friendly CSV and take a look at the data so far
csv.writer(open('/home/vagrant/repos/datasets/preliminary_artist_table.csv', 'wb')).writerows(artist_table_rows)
df = pd.read_csv("/home/vagrant/repos/datasets/preliminary_artist_table.csv", header = None)
df.describe
```

```python
#Just a list of the artists
artists = []
for house_sg in list_of_house_subgenres:
    sg_artist_dict = en.get('genre/artists', name=house_sg)
    for a in sg_artist_dict['artists']:
        sg_row = (a["name"].encode('utf-8'))
        artists.append(sg_row)
```

```python
artists[0:20]
```

```python
start_time = time.time()
#^^^Time this whole thing
#Table of songs built on the previous artist table
#The structure of this table is artist, song, song ID
song_table_rows = []
for a in artists:
    playlist_dict = en.get('playlist/static', artist= a , type='artist-radio' )
    for x in playlist_dict[u'songs']:
        playlist_row =(a , x['title'].encode('utf-8'), x['id'].encode('utf-8'))
        song_table_rows.append(playlist_row)
end_time = time.time()
completion_time = end_time - start_time
print "This took {} seconds to run".format(completion_time)
#take a peek at this data
song_table_rows[0:50]
```

```python
#let's make it a CSV so we don't have to recall this data ever again
csv.writer(open('/home/vagrant/repos/datasets/all_the_songs.csv', 'wb')).writerows(song_table_rows)
df = pd.read_csv("/home/vagrant/repos/datasets/all_the_songs.csv", header = None)
```
