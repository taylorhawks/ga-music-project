```python
import numpy as np
from scipy.spatial.distance import sqeuclidean

from collections import Counter
from sklearn.metrics.cluster.unsupervised import _intra_cluster_distance

def get_centroid(i, df, kmeans):
    """
    Compute the centroid (center of mass) of cluster i.
    """
    df_copy = df.copy()
    # add "label" as a column so we can filter on label
    df_copy["label"] = kmeans.labels_

    # filter just for the points in cluster i
    points_in_cluster_i = df_copy[df_copy["label"] == i]

    # remove the extra column called "label"
    del points_in_cluster_i["label"]
    
    # compute the average
    return np.mean(points_in_cluster_i)
    
def intra_cluster_variation(kmeans, df):
    """
    kmeans: KMeans object instantiated with `n_clusters` and `n_init` and fitted to the data    

    This returns B = sum((1/p) * (distance center of cluster to center of data )^2 for all the clusters),
    where p is the number of points in each cluster.
    """
    # group the labels by the centers
    label_counter = Counter(kmeans.labels_)
    average_of_all_the_data = np.mean(df)
    intra_cluster_variations = []

    for i, label in enumerate(label_counter.items()):
        # this is p
        number_of_points_in_cluster = label_counter[i]
        centroid = get_centroid(i, df, kmeans)
        variation = (1 / float(number_of_points_in_cluster)) * (sqeuclidean(average_of_all_the_data, centroid))
        intra_cluster_variations.append(variation)
    return sum(intra_cluster_variations)
```

```python
def ch_index(kmeans, df):
    b = intra_cluster_variation(kmeans, df)
    n = float(len(df))
    k = float(kmeans.n_clusters)
    w = kmeans.inertia_
    return (w/ (k-1)) / (b / (n-k))
```

```python
ch_scores=[]
clusterz = range(2,30)
for i in clusterz:
    kmeans = KMeans(n_init=10, n_clusters = i)
    kmeans.fit(df)
    ch_scores.append(ch_index(kmeans, df))
plt.plot(clusterz, ch_scores)
```
